```python
import datetime

hour_counts = {}
with open('Testorder_data.csv', 'r') as file:
    orders = []
    for line in file:
        values = line.strip().split('\t')
        order_id = values[0]
        driver_id = values[1]
        passenger_id = values[2]
        
        removee=values[3].replace('"', '')
        start_region_hash = removee
        dest_region_hash = values[4]
        price = values[5]
        removee1=values[6].replace('"', '')
        order_time=removee1
        #time stamp extracting max order hours
        order_hour = datetime.datetime.strptime(order_time, '%Y-%m-%d %H:%M:%S').hour
        
        # Increment the count for the corresponding hour
        if order_hour in hour_counts:
            hour_counts[order_hour] += 1
        else:
            hour_counts[order_hour] = 1
        orders.append([order_id, driver_id, passenger_id, start_region_hash, dest_region_hash, price,order_time])  
# Print the counts for each hour
for hour, count in hour_counts.items():
    print(f"Hour {hour}: {count} orders")
       
    
        
```

```python

# In[14]:
# Create a PrettyTable object
table = PrettyTable(['Field', 'Type', 'Meaning', 'Example'])
for order in orders:
    table.add_row(['order_id', 'string', 'order ID', order[0]])
    table.add_row(['driver_id', 'string', 'driver ID', order[1]])
    table.add_row(['passenger_id', 'string', 'user ID', order[2]])
    table.add_row(['start_region_hash', 'string', 'departure', order[3]])
    table.add_row(['dest_region_hash', 'string', 'destination', order[4]])
    table.add_row(['price', 'double', 'price', order[5]])

# Print the table
print(table)


# In[ ]:




```

```python
from collections import Counter
with open('Testpoi_data.csv', 'r') as file:
    poi = []
    for line in file:
        values = line.strip().split('\t')
        remove3=values[0].replace('"', '')
        region_hash = remove3
        poi_class = values[1]
        poi.append([region_hash, poi_class])
        
#number of facilities in region
region_facilities = Counter([p[0] for p in poi])

number_of_facilities_per_region=dict(region_facilities)
print(number_of_facilities_per_region)
#number of specific facilities in region
# Create a PrettyTable object
table = PrettyTable(['Field', 'Type', 'Meaning', 'Example'])
for p in poi:
    table.add_row(['region_hash', 'string', 'Region hash', p[0]])
    table.add_row(['poi_class', 'string', 'POI class and its number', p[1]])

# Print the table
print(table)
# Initialize an empty dictionary
facility_counts = {}

# Iterate over each poi record
for p in poi:
    region_hash = p[0]
    poi_class = p[1]
    
    # Parse the poi_class field to extract the facility type and its count
    if '#' in poi_class:
        facility_type, count = poi_class.split(':')
        level1, level2 = facility_type.split('#')
        
    
    
    # Update the facility_counts dictionary with the count for this facility type in this region
    if region_hash not in facility_counts:
        facility_counts[region_hash] = {}
    facility_counts[region_hash][level1] = facility_counts[region_hash].get(level1, 0) + int(count)
print(facility_counts)
```
